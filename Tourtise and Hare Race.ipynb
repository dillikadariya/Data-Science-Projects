## Simulation of Tourtise and Hare Race

```python
import random

print('ON YOUR MARK, GET SET \n  BANG !!!!! \n AND THEY ARE OFF !!!!!')
def main():
    # Initializing variables
    tort_position = 1
    hare_position = 1
   
    while tort_position < 70 and hare_position < 70:
        # Calculating move distances for the tortoise and the hare
        tortoise_move_distance = random.randint(1, 10)
        hare_move_distance = random.randint(1, 10)
        
        # Updating positions based on move distances
        tort_position = tortoise_move(tort_position, tortoise_move_distance)
        hare_position = hare_move(hare_position, hare_move_distance)
        
        # Print positions
        print(f'Tortoise position: {tort_position}, Hare position: {hare_position}')
        
    if tort_position >= 70 and hare_position < 70:
        print('TORTOISE WINS!!! YAY!!!')
    elif hare_position >= 70 and tort_position < 70:
        print('HARE WINS. Yach!')
    else:
        print('It is a tie')

def hare_move(hare_position, move_distance):
    if move_distance < 3:
        hare_position += 0
    elif move_distance < 5:
        hare_position += 9
    elif move_distance == 5:
        hare_position -= 12
    elif move_distance < 8:
        hare_position += 1
    elif move_distance <= 10:
        hare_position -= 2
    else:
        print('Error')
    
    # hare's position is not negative
    if hare_position < 1:
        hare_position = 1
    
    return hare_position

def tortoise_move(tort_position, move_distance):
    if move_distance < 6:
        tort_position += 3
    elif move_distance < 8:
        tort_position -= 6
    elif move_distance < 11:
        tort_position += 1
    else:
        print('Error')
    
    # tortoise's position is not negative
    if tort_position < 1:
        tort_position = 1
    
    return tort_position

if __name__ == "__main__":
    main()
```
