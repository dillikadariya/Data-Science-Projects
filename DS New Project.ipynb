## #3 Visualizing the Data

### A. This method produces a bar graph for the 'Expenditure' for each country.


```python
import matplotlib.pyplot as plt

# Sorting the data by Expenditure for better visualization
sorted_data = WHOdata.sort_values(by='Expenditure', ascending=True)

# Plotting the bar graph
plt.figure(figsize=(12, 8))  # Increase figure size
plt.bar(sorted_data['Country'], sorted_data['Expenditure'], color='skyblue') # axis and color of bars
plt.xlabel('Country')
plt.ylabel('Expenditure')
plt.title('Expenditure by Country')
plt.xticks(rotation=45, ha='right')  # Rotates x-axis labels and align to the right
plt.subplots_adjust(bottom=0.4) # Adjusts bottom spacing
plt.tight_layout()
plt.show()
```

### B. Scatterplot with country serial number on X-axis and mental health expenditure on Y-axis.

### Scatterplot with country names on X-axis looks illegible, so this code creates a new column "S.N" that provides a serial number to each countries that are sorted based on their expenditure on mental health.

```python
df = pd.DataFrame(sorted_data)
df['S.N'] = range(1, 79)    # creates new column ranging 1 to 78
indexed_data = sorted_data[['S.N','Country','Expenditure']].sort_values(by='Expenditure', ascending=True)
indexed_data
```

### Scatter plot based on their S.N

```python
import pandas as pd
import matplotlib.pyplot as plt

# Plots 'Country S.N' on the x-axis and 'Expenditure' on the x-axis as dots
df.plot( x='S.N', y='Expenditure', style='o', figsize=(11, 6))

plt.xlabel('Country Serial Nmmber')
plt.ylabel('Expenditure')
plt.title('Expenditure by Country')
plt.grid(True)
plt.xticks(range(1, 79))   # Sets x-axis ticks from 1 to 78
plt.locator_params(axis='x', tight=True, nbins=25) # adjusts space between x values
plt.show()
```

## C. Top 10 countries that invested the most on mental health

### This code defines a function "top" that returns the n rows with the highest value for the specified column. The top accepts a dataframe as its first input, a parameter named n that accepts a number and provides a reasonable default, and a parameter called column that defaults to 'Expenditure'.

```python
def top(df, n=5, column='Expenditure'):
    return df.sort_values(by=column, ascending=False)[:n]
top(Data, n=10)   # displays highest 10 values
```

## D. 10 countries that invested the least on mental health

### This code defines a function "bottom" that returns the n rows with the lowest value for the specified column. The top accepts a dataframe as its first input, a parameter named n that accepts a number and provides a reasonable default, and a parameter called column that defaults to 'Expenditure'.

```python
def bottom(df, n=5, column='Expenditure'):
    return df.sort_values(by=column, ascending=True)[:n]
```

```python
 bottom(Data, n=10) # lowest 10 values
```
